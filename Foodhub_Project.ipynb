```python
"""
Conditions:
- restaurants must have rating count of more 50
- must also have average rating greater than 4
"""

#disregard 0 ratings
df0 = df[df["rating"] != 0]

#groupby -> groups items in dataFrame by a particular catergory (a)
#grouby restaurant name means all rows with same restaurant name will be grouped together

"""
agg is aggregation operation. here using agg, you're summing up values in each group
other agg functions
- count, sum, mean, min/max

here the rating: sum part -> calculates sum of ratings per restaurant/all ratings
of a particular restaurant will be addded together

RENAME
- this function is used to change the name of the resulting column
- the rating column is now renamed to r_sum bc it nows contains the sum of rating
  for each restaurant
"""
df0 = df0.groupby("restaurant_name").agg({"rating": "sum"}).rename(columns = {"rating": "r_sum"})

#calculate average

"""
-> single brackets [] returns series
-> double brackets [[]] returns dataFrames
df[["restaurant_name", "rating"]] -> selects only those 2 columns from df dataFrame

ANOTHER WAY
df[["restaurant_name", "rating"]][df["rating"]!= 0] -> df[["restaurant_name]][df["rating"] !=0]

-> the outer brackets here: [df["rating"]!= 0]
- applies the bool condition to the dataFrame, meaning only rows where rating != 0 will be selected

mean: applies to ALL numerical types
-after groupping it gets avg of rating for each restaurant
- the >4 part compares avg for each restaurant to 4, returning True or False
- mean function is applied to rating column bc its the only numerical type avaiable


"""

dfmean = df[["restaurant_name", "rating"]][df["rating"]!= 0].groupby("restaurant_name").mean()>4

#replaces values in dfmean with only rows where rating is True
dfmean = dfmean[dfmean["rating"] == True]

#now you want the count of ratings to be more than 50
dfcount = df[["restaurant_name", "rating"]][df["rating"] != 0].groupby("restaurant_name").count()> 50

#make sure dataFrame ONLY has restaurants with count more than 50
#in the dfcount dataFrame, bool comparison for rating column has to be True
dfcount = dfcount[dfcount["rating"] == True]


"""
pd.merge() is pands function that combines 2 dataFrames based on common column or index
- here, dfmean and dfcount have common coolumn called restaurant name
- "on" specifies what the merging should be done on. here its done on restaurant name
-> means: only rows where restaurant_name exists in both dataFrames will be merged together
-> 'how' parameter specifies inner join
-> inner join means only rows with matching values in restaurant_name column in both dataFrames will be included
-> ex: if restaurant name in dfmean not in dfcount, that row will be exlcuded & vice versa

dflist is the new dataFrame with the merged dataFrames
"""

dflist = pd.merge(dfmean, dfcount, on = "restaurant_name", how = "inner")

dflist
```


restaurant 25% on the orders having cost greater than 20 dollars and 15% on the orders having cost greater than 5 dollars. Add net revenue generated by the company across all orders.

meaning:
- For orders where the cost of the order is greater than 20 dollars, the company takes 25 percent of the total order cost as its revenue.
- For orders where the cost is greater than $5 but less than or equal to 20 dollars, the company takes 15 percent of the total order cost as its revenue.

```python
df_price = df["cost_of_the_order"] #only put cost of order column into new dataFrame

net_revenue = 0

for i in range(len(df_price)):
  if(df_price[i] > 20): #orders greater than 20
    net_revenue = net_revenue + df_price[i]*0.25
  elif (df_price[i] > 5 and df_price[i] <= 20): #orders greater than 5 and <= 20
    net_revenue += df_price[i]*0.15 #updating total/net revenue
print(f"The total revenue is:{net_revenue:.2f}. ")



```

```python
#Total time required to deliver the food. percentage of orders take more than 60 minutes to get delivered from the time the order is placed.
#The food has to be prepared and then delivered.food prep time and delivery time both need to be considered
df['total_time'] = df['food_preparation_time'] + df['delivery_time']

#we need to calculate a fraction: orders over 60 percent/total num of orders

totalOrders = len(df[df['total_time'] >float(0)]) #can be len(df)
#this is bc time is positive anyways, so comparison is not needed

orderOver60 = len(df[df['total_time'] > float(60)])

percent = (float(orderOver60))/float(totalOrders)
percent = percent*100

print(f"The percent is: {percent:.2f} %")
```

```python

"""
the delivery time of the orders on weekdays and weekends. Mean delivery time vary during weekdays and weekends
1. groupby day of the week: this separates weekdays and weekends
-> forms serparate groups
-> goal is to perform an operation on each group (mean operation)
2. the agg function allows you to apply a function to EACH group (IMPORTANT)
-> here you are applying mean function to every group in day_of_the_week (weekdays and ends)
3. so delivery_time:mean
-> telling pandas to compute mean of delivery time column for reach group in the dataFrame...
-> ... AFTER it has been grouped
"""
deliveryTimeAvg = df.groupby("day_of_the_week").agg({"delivery_time": "mean"})
print(deliveryTimeAvg)
```
